# multiple plot
1.Subplot:  syntax= subplot(row,columns,index)

```python
# 1 row 2 columns
plt.subplot(1,2,1)   # 1st plot
plt.subplot(1,2,2)   # 2nd plot
plt.show()
```

```python
#2 row 1 columns
plt.subplot(2,1,1)
plt.subplot(2,1,2)
plt.show()
```

```python
#2 row 2 columns
plt.subplot(2,2,1)   # subplot 1
plt.subplot(2,2,2)   # subplot 2
plt.subplot(2,2,3)   # subplot 3
plt.subplot(2,2,4)   # subplot 4
plt.show()
```

# 2. add_subplot
syntax=add_subplot(row,col,index)  1. figure represent entire area where data visualization will be displayed   2. axes is like and individual plot or chart within figure


```python
#add subplot to the figure
fig=plt.figure()
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
```

# create multiple axes with examples

```python
#create data
x=[0,1,2,3]
y=[3,8,1,10]
```

```python
fig,ax=plt.subplots(1,2)
ax[0].plot(x,y)
ax[1].bar(x,y)
fig.suptitle("figure of 2 plot")
plt.show()
```

# Alternate code

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.plot(x,y)
ax2.bar(x,y)
plt.suptitle(" fig ax1 ax2")
plt.show()
```

# vertically stacked subplots

```python
fig,ax=plt.subplots(2)  # row=0,col=2
ax[0].plot(x,y)
ax[1].bar(x,y)
plt.show()
```

```python
#2 rows and 2 columns
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].plot(x,y,"tab:orange")
ax[1,0].plot(x,y,"tab:red")
ax[1,1].plot(x,y,"tab:green")
plt.show()
```

```python
#Alternate code
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
ax1.plot(x,y)
ax2.plot(x,y,"tab:orange")
ax3.plot(x,y,"tab:red")
ax4.plot(x,y,"tab:green")
plt.show()
```

```python
#legend
plt.plot(x,y)
plt.plot(y,x,"r")
plt.legend(["dataset1","dataset2"])
plt.show()
```

# more diffrent types of plots
Bar graph bar()

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
#change colors of bar
c=["r","b","g","y"]
plt.bar(x,y,color=c)
plt.show()
```

# horizontal bar graph

```python
plt.barh(x,y)
plt.show()
```

```python
#change width of bar
#width use in vertical
plt.bar(x,y,width =0.3)
plt.show()
```

```python
#height use in horizontal
plt.barh(x,y,height=0.2,color="r")
plt.show()
```

# scatter plot

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.scatter(x,y,color="y",s=200) # s is size of dot
plt.show()
```

# pie chart

```python
x=[10,20,30,40]
y=["a","b","c","d"]    #create labels
plt.pie(x,labels=y)
plt.show()
```

# histogram

```python
plt.hist([1,2,2,3,4,4,5],bins=5)
plt.title(" histogram")
plt.xlabel("values")
plt.ylabel("frequency")
plt.show()
```

# 3d plots

# 3d  line plot

```python
#creating data
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,8,27,64,125]
```

```python
fig=plt.figure()
#creates 3d plot
ax=plt.axes(projection="3d")
ax.plot3D(z,y,x)
plt.show()
```
